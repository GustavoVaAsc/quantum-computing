```python
# 1.1
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex

print("Cubit recién inicializado:")
circuit = QuantumCircuit(1)
state_vector = Statevector(circuit)
display(array_to_latex(state_vector.data, prefix="|ψ⟩ = "))
display(state_vector.draw("latex"))

circuit.x(0)
print("Después de aplicar la compuerta X:")
state_vector = Statevector(circuit)
display(array_to_latex(state_vector.data, prefix="|ψ⟩ = "))
display(state_vector.draw("latex"))

circuit.h(0)
print("Después de aplicar la compuerta de Hadamard:")
state_vector = Statevector(circuit)
display(array_to_latex(state_vector.data, prefix="|ψ⟩ = "))
display(state_vector.draw("latex"))

circuit.y(0)
print("Después de aplicar la compuerta Y:")
state_vector = Statevector(circuit)
display(array_to_latex(state_vector.data, prefix="|ψ⟩ = "))
display(state_vector.draw("latex"))

circuit.z(0)
print("Después de aplicar la compuerta Z:")
state_vector = Statevector(circuit)
display(array_to_latex(state_vector.data, prefix="|ψ⟩ = "))
display(state_vector.draw("latex"))

circuit.s(0)
print("Después de aplicar la compuerta S:")
state_vector = Statevector(circuit)
display(array_to_latex(state_vector.data, prefix="|ψ⟩ = "))
display(state_vector.draw("latex"))

circuit.t(0)
print("Después de aplicar la compuerta T:")
state_vector = Statevector(circuit)
display(array_to_latex(state_vector.data, prefix="|ψ⟩ = "))
display(state_vector.draw("latex"))
```

# 1.2

- El cubit se inicializa en el estado $\ket{0}$
- Después se le aplica la compuerta $X$, que es el equivalente cuántico a la compuerta NOT, por lo tanto ahora el estado es $\ket{1}$
- Posterior a $X$ se le aplica la compuerta de Hadamard $H$, convirtiendo el estado $\ket{1}$ en $\ket{-} = \frac{1}{\sqrt{2}}(\ket{0} - \ket{1})$
- Después de aplicar a $H$, se le aplica la compuerta $Y$, la cual invierte el signo del componente $\ket{1}$ y multiplica por la unidad imaginaria $i$ ambas componentes. Esto representa una rotación de $180 \degree$ en el eje y de la esfera de Bloch
- Posterior a aplicar $Y$ se aplica la compuerta $Z$, que invierte el signo del componente $\ket{1}$, pero esta vez sin multiplicar por la unidad imaginaria. Esta compuerta representa una rotación de $180 \degree$ en la esfera de Bloch.
- Una vez aplicada la compuerta $Z$, aplicamos la compuerta $S$, que se puede interpretar "la raíz cuadrada de $Z$", es decir, aplica una rotación de $90 \degree$ en el eje $Z$ de la esfera de Bloch, por lo tanto, multiplica por la unidad imaginaria el componente $\ket{1}$.
- Finalmente aplicamos $T$, que se puede interpretar como "la raíz cuadrada de $S$", es decir, aplica una rotación de $45 \degree$ en el eje $Z$ de la esfera de Bloch, en concreto, multiplica por un factor de $e^{\frac{\pi}{4}i}$ el componente $\ket{1}$

$$\ket{\psi_0} = \myvector{1 \\ 0}$$

$$\ket{\psi_1} = X \ket{\psi_0} = \mymatrix{cc}{0 & 1 \\ 1 & 0} \myvector{1 \\ 0}$$

$$\ket{\psi_1} = \myvector{(0)(1) + (1)(0) \\ (1)(1) + (0)(0)} = \myvector{0 \\ 1}$$

$$\ket{\psi_2} = H \ket{\psi_1} = \frac{1}{\sqrt{2}} \mymatrix{cc}{1 & 1 \\ 1 & -1} \myvector{0 \\ 1}$$

$$\ket{\psi_2} = \frac{1}{\sqrt{2}} \myvector{(1)(0) + (1)(1) \\ (1)(0) + (-1)(1)} = \myvector{\frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}}}$$

$$\ket{\psi_3} = Y \ket{\psi_2} = \mymatrix{cc}{0 & -i \\ i & 0} \myvector{\frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}}}$$

$$\ket{\psi_3} = \myvector{(0)(\frac{1}{\sqrt{2}}) + (-i)(-\frac{1}{\sqrt{2}}) \\ (i)(\frac{1}{\sqrt{2}}) + (0)(-\frac{1}{\sqrt{2}})} = \myvector{\frac{i}{\sqrt{2}} \\ \frac{i}{\sqrt{2}}}$$

$$\ket{\psi_4} = Z \ket{\psi_3} = \mymatrix{cc}{1 & 0 \\ 0 & -1} \myvector{\frac{i}{\sqrt{2}} \\ \frac{i}{\sqrt{2}}}$$

$$\ket{\psi_4} = \myvector{(1)(\frac{i}{\sqrt{2}}) + (0)(\frac{i}{\sqrt{2}}) \\ (0)(\frac{i}{\sqrt{2}}) + (-1)(\frac{i}{\sqrt{2}})} = \myvector{\frac{i}{\sqrt{2}} \\ -\frac{i}{\sqrt{2}}}$$

$$\ket{\psi_5} = S \ket{\psi_4} = \mymatrix{cc}{1 & 0 \\ 0 & i} \myvector{\frac{i}{\sqrt{2}} \\ -\frac{i}{\sqrt{2}}}$$

$$\ket{\psi_5} = \myvector{(1)(\frac{i}{\sqrt{2}}) + (0)(-\frac{i}{\sqrt{2}}) \\ (0)(\frac{i}{\sqrt{2}}) + (i)(-\frac{i}{\sqrt{2}})} = \myvector{\frac{i}{\sqrt{2}} \\ \frac{1}{\sqrt{2}}}$$

$$\ket{\psi_6} = T \ket{\psi_5} = \mymatrix{cc}{1 & 0 \\ 0 & e^{i\pi/4}} \myvector{\frac{i}{\sqrt{2}} \\ \frac{1}{\sqrt{2}}}$$

$$\ket{\psi_6} = \myvector{(1)(\frac{i}{\sqrt{2}}) + (0)(\frac{1}{\sqrt{2}}) \\ (0)(\frac{i}{\sqrt{2}}) + (e^{i\pi/4})(\frac{1}{\sqrt{2}})} = \myvector{\frac{i}{\sqrt{2}} \\ \frac{e^{i\pi/4}}{\sqrt{2}}}$$
